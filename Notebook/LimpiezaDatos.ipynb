```python
# Limpieza y unificación de datos
import pandas as pd

# Cargar base original
df = pd.read_excel("BD_San_Lorenzo_2020-2024.xlsx")

# Revisar valores nulos
print(df.isnull().sum())

# Eliminar duplicados
df = df.drop_duplicates()

# Rellenar valores faltantes (ejemplo con promedio)
df = df.fillna(df.mean(numeric_only=True))

# Homogenizar fechas
df['fecha'] = pd.to_datetime(df['fecha'])

# Guardar base limpia
df.to_excel("base_limpia.xlsx", index=False)
print("Base limpia guardada.")
```
